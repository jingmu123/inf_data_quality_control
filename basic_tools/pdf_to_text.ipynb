```python
import requests
import json
import glob
import os
```

```python

# 上传单个文件
def upload_single_file(file_path, url):
    with open(file_path, 'rb') as file:
        files = {'file': file}
        response = requests.post(url, files=files, data = {'params_str' : json.dumps({'language':'ch'})})
        basename = os.path.basename(file_path).replace('.png', '.md')
        with open(f'pngs/{basename}', 'w') as md_file:
            md_file.write(response.json()['data']['content'])
        return response.json()['data']['content']

```

```python
data_path = "../../full_data/nihaixia/【视频同步文稿】倪海厦人纪系列书籍之《皇帝内经上册》 .pdf"
url = 'http://a21.infly.cloud:10087/convert_file'
response = upload_single_file(data_path,url)
```
